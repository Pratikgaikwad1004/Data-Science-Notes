```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
y
```

```python
x
```

```python
# Functional
plt.plot(x,y)
plt.xlabel("X LABLE")
plt.ylabel("Y LABLE")
plt.title("TITLE")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Orainted Method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Lable')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes2.plot(y, x)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title("First plot")

axes[1].plot(y, x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

# Figure Size and DPI

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 3))
axes[0].plot(x, y)
axes[0].set_title("First plot")

axes[1].plot(y, x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

```python
# Save figure
fig.savefig("My_Fig.png", dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label='Square')
ax.plot(x, x**3, label='cube')

ax.legend()
```

# Plot Apperarance

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="red", linewidth=2, alpha=1, linestyle='-', marker='o', markersize=20, 
        markerfacecolor='yellow') # We can use RGB Hex code
```
